## Overview

Microsoft wants to start a movie studio, but they don't know what kind of movie they should make or who they should hire to work on it. IMDB is a website that contains a lot of information on movies and produces many CSV's for us to glean information about what makes a movie successful. After parsing through the data, it seems that around 10 directors have experience making multiple very popular films, the genre should combine action and adventure and likely Sci-Fi or Fantasy, and Microsoft should spend between 50 and 300 million dollars for their production budget.

## Business Problem

Microsoft is starting a movie studio, but doesn't have experience in show business, so I gathered data to make recommendations for their first movie.

My main questions are:
* Which directors have the most experience directing high grossing movies?

```python
budg_int['production_budget'] = budg_int['production_budget'].str.replace(',', '')
budg_int['production_budget'] = budg_int['production_budget'].str.replace('$', '')
budg_int['production_budget'] = budg_int['production_budget'].astype(int)
budg_int.head()
```
